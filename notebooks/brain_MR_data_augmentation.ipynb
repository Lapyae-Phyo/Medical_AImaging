## 1.3. Data Augmentation

### 1.3.1 Pituatary Lesion Augmentation

```python
# Define the path to the directory containing your original images
original_dataset_path = '../data/MR/brain/pituatary_lesion/'
```

```python
# Define the path to the directory where augmented images will be saved
augmented_dataset_path = '../data/MR/brain/pituatary_lesion_augmented/'
```

```python
# Ensure the augmented dataset directory exists, create if not
os.makedirs(augmented_dataset_path, exist_ok=True)
```

```python
# Create an instance of the ImageDataGenerator with desired augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)
```

```python
# Get the list of image files in the original dataset directory
image_files = [f for f in os.listdir(original_dataset_path) if os.path.isfile(os.path.join(original_dataset_path, f))]
```

```python
# Specify the number of augmented images you want per original image
augmentation_factor = 7  # Adjust as needed to reach your target number (83 * 7 = 581)
```

```python
# Apply data augmentation and save the augmented images
for image_file in image_files:
    img_path = os.path.join(original_dataset_path, image_file)
    img = image.load_img(img_path)
    img_array = image.img_to_array(img)
    img_array = img_array.reshape((1,) + img_array.shape)

    i = 0
    for batch in datagen.flow(img_array, batch_size=1, save_to_dir=augmented_dataset_path, save_prefix='aug', save_format='jpg'):
        i += 1
        if i >= augmentation_factor:
            break
```

```python
# Display a few augmented images for visualization
sample_augmented_images = [f for f in os.listdir(augmented_dataset_path) if f.startswith('aug')][:5]
```

```python
plt.figure(figsize=(10, 5))
for i, img_name in enumerate(sample_augmented_images):
    img = image.load_img(os.path.join(augmented_dataset_path, img_name))
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.axis('off')

plt.show()
```
